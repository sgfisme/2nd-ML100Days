```python
import os
import numpy as np
import pandas as pd
```

```python
# 設定 data_path
dir_data = '../data/Part01/'
```

```python
f_app = os.path.join(dir_data, 'application_train.csv')
print('Path of read in data: %s' % (f_app))
app_train = pd.read_csv(f_app)
```

### 如果沒有想法，可以先嘗試找出剛剛例子中提到的問題的答案
#### 資料的 row 數以及 column 數

```python
# 資料的 row 數
print("資料的 row 數: ", len(app_train))

# 資料的 column 數
print("資料的 column 數: ", app_train.shape[1])
```

#### 列出所有欄位

```python
app_train.head(5)
```

#### 截取部分資料

```python
# 截取 SK_ID_CURR 資料
app_train['SK_ID_CURR'].head(5)
```

```python
# 截取 SK_ID_CURR, NAME_CONTRACT_TYPE, AMT_INCOME_TOTAL
app_train[['SK_ID_CURR', 'NAME_CONTRACT_TYPE', 'AMT_INCOME_TOTAL']].head(10)
```

```python
# 截取第 5 筆到第 10 筆中的 SK_ID_CURR, NAME_CONTRACT_TYPE, AMT_INCOME_TOTAL 欄位
app_train.loc[5:10, ['SK_ID_CURR', 'NAME_CONTRACT_TYPE', 'AMT_INCOME_TOTAL']]
```

#### 還有各種數之不盡的資料操作，重點還是取決於實務中遇到的狀況和你想問的問題，在馬拉松中我們也會陸續提到更多例子

```python
# 篩選 CNT_CHILDREN > 2 的資料
app_train[app_train['CNT_CHILDREN'] > 2]
```

```python
# 依 CODE_GENDER 加總 AMT_INCOME_TOTAL, AMT_CREDIT, AMT_ANNUITY, AMT_GOODS_PRICE
app_train[['CODE_GENDER','AMT_INCOME_TOTAL','AMT_CREDIT','AMT_ANNUITY','AMT_GOODS_PRICE']].groupby(by='CODE_GENDER').sum()
```

```python
# 依 CODE_GENDER 針對 AMT_INCOME_TOTAL, AMT_CREDIT, AMT_ANNUITY, AMT_GOODS_PRICE 取平均數
app_train[['CODE_GENDER','AMT_INCOME_TOTAL','AMT_CREDIT','AMT_ANNUITY','AMT_GOODS_PRICE']].groupby(by='CODE_GENDER').mean()
```
